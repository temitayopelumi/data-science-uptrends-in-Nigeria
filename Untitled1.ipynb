```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Mcr2017=pd.read_csv('multipleChoiceResponses.csv',  encoding='latin-1')
```

```python
Mcr2018=pd.read_csv('archive (1)/multipleChoiceResponses.csv',  encoding='latin-1')
```

```python
Mcr2019= pd.read_csv('multiple_choice_responses.csv')
```

```python
Mcr2020 = pd.read_csv('kaggle_survey_2020_responses.csv')
```

```python
Nigeria2017 = Mcr2017[Mcr2017['Country']=='Nigeria']
Nigeria2018 = Mcr2018[Mcr2018['Q3']=='Nigeria']
Nigeria2019 = Mcr2019[Mcr2019['Q3']=='Nigeria']
Nigeria2020 = Mcr2020[Mcr2020['Q3']=='Nigeria']
others2017 = Mcr2017[Mcr2017['Country'] != 'Nigeria']
others2018 = Mcr2018[Mcr2018['Q3'] != 'Nigeria']
others2019 = Mcr2019[Mcr2019['Q3'] != 'Nigeria']
others2020 = Mcr2020[Mcr2020['Q3']!='Nigeria']
```

Data science usage in Nigeria

```python
nigeriaCount = [len(Nigeria2017),len(Nigeria2018),len(Nigeria2019),len(Nigeria2020)]
othersCount = [len(others2017),len(others2018),len(others2019),len(others2020)]
countDict = {"Nigeria":nigeriaCount, "Rest of the World": othersCount, "Year": ['2017','2018','2019','2020']}
countDataframe = pd.DataFrame(countDict) 
```

```python
countDataframe
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
countDataframe.plot(x='Year', y='Nigeria', ax=ax1, color='mediumseagreen') # Our Nigerian sample will be plotted with green
countDataframe.plot(x='Year', y='Rest of the World', ax=ax2)
fig.suptitle('Comparing Number of Survey Respondents over the years')

plt.show()
```
